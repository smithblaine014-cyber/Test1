Program and apply 1 inch of irrigation whenever the cum_ET is 1 inch or greater than the cum_P_and_I  whenever in the season. Recall the if statement and for loop we practiced earlier.


Calculate a new variable cum_P_and_I as the cumulative sum of precipitation P and irrigation I.


```python
# ==================================
# RELOAD DATA (RUN THIS FIRST)
# ==================================
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt

# Get the filename from the uploaded dictionary (assuming a single file is uploaded)
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Create DOY
df["DOY"] = range(1, len(df) + 1)

# Create cumulative variables
df["cum_P"] = df["Precipitation_inches"].cumsum()
df["cum_ET"] = df["ET_inches"].cumsum()

print("Data loaded successfully.")
print(df.head())
```

```python
# =====================================================
# SECTION 3 – BASIC IRRIGATION SCHEDULING
# =====================================================

# Ensure cumulative ET and cumulative precipitation columns exist.
# These are crucial for the irrigation logic. If they haven't been created yet,
# this step re-creates them based on 'Precipitation_inches' and 'ET_inches'.
df["cum_P"] = df["Precipitation_inches"].cumsum()
df["cum_ET"] = df["ET_inches"].cumsum()

# -----------------------------------------------------
# 1. Initialize irrigation columns
# -----------------------------------------------------
# 'I' will store the daily irrigation amount (in inches).
# 'cum_P_and_I' will store the cumulative sum of precipitation and irrigation.
df["I"] = 0.0   # Initialize daily irrigation to 0 for all days
df["cum_P_and_I"] = 0.0 # Initialize cumulative P + I to 0

# -----------------------------------------------------
# 2. Loop through the DataFrame to simulate irrigation day by day
# -----------------------------------------------------

for i in range(len(df)):

    # Condition 1: Only apply irrigation during the defined growing season (May to October).
    # This filters out days outside the typical crop growth period.
    if 5 <= df.loc[i, "Month"] <= 10:

        # Condition 2: Check the irrigation trigger.
        # Irrigation is applied if cumulative ET (water demand) is 1 inch or more
        # greater than the cumulative water supply (precipitation + previous irrigation).
        if df.loc[i, "cum_ET"] >= df.loc[i, "cum_P_and_I"] + 1:

            df.loc[i, "I"] = 1.0  # Apply 1 inch of irrigation if conditions are met

    # Update cumulative precipitation + irrigation for the current day.
    # This calculation considers the precipitation and any irrigation applied on the current day.
    if i == 0: # For the first day
        df.loc[i, "cum_P_and_I"] = df.loc[i, "Precipitation_inches"] + df.loc[i, "I"]
    else: # For subsequent days, add to the previous day's cumulative total
        df.loc[i, "cum_P_and_I"] = (
            df.loc[i-1, "cum_P_and_I"]
            + df.loc[i, "Precipitation_inches"]
            + df.loc[i, "I"]
        )

# -----------------------------------------------------
# FIGURE 6 – Plotting Cumulative ET vs Cumulative P + I
# -----------------------------------------------------
# This plot visually compares the water demand (cum_ET) against the water supply
# from both precipitation and irrigation (cum_P_and_I).
plt.figure()
plt.plot(df["DOY"], df["cum_ET"], label="Cumulative ET", color='red')
plt.plot(df["DOY"], df["cum_P_and_I"], label="Cumulative Precipitation + Irrigation", color='blue')
plt.xlabel("Day of Year")
plt.ylabel("Cumulative Depth (Inch)")
plt.title("Figure 6: Cumulative ET vs Cumulative P + I")
plt.legend()
plt.grid(True) # Add a grid for better readability
plt.show()

# Show total irrigation applied over the entire period.
print("Total irrigation applied during season (inches):", df["I"].sum())
```
